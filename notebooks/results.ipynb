```python
import wandb

api = wandb.Api()

wandb.login()
```

```python
project = api.project("rf-model-7-b72b89f1")
```

```python
print(project.entity)
```

```python
projects = api.projects("philip-baumann-hslu")
```

```python
projects_rf_7s = [p.name for p in projects if p.name.__contains__("-7-") and p.name.__contains__("rf-model")]
```

```python
import json
completed_runs = []
for p in projects_rf_7s:
    runs = api.runs(f"{projects.entity}/{p}",  filters={"state": "finished"})
    for run in runs:
        metrics = json.loads(run.summary_metrics)
        metrics["name"] = run.name
        metrics["feature_set"] = p
        completed_runs.append(metrics)

print(completed_runs)
```

```python
completed_runs
```

```python
completed_runs[0]
```

```python
filtered_runs = [r for r in completed_runs if r["accuracy"] > 0.6 and r["logloss"] < 0.6]
```

```python
[f"{r["name"]}, {r["feature_set"]}:acc{r["accuracy"]}, {r["logloss"]}" for r in filtered_runs]
```
